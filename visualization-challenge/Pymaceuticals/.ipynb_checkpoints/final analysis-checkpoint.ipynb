### Analysis
- We run a drug test, on 250 subjects with initial tumor size of 45mm3, of 10 drugs. Capomulin and Ramicane showed measurable improvement in the tumor size, while the rest showed little or negligible impact on the tumor, which also increased in size during treatment
### Conclusions
- There are only two effective drug treatments that showed tumor reduction (Capomulin and Ramicane). Both had similar performance. All other drugs either had no impact or all performance the same and the tunor increased in size
- The average final tumor volume is linearly dependent to the weight of the mouse for effective drug treatments, over the weight range of the study (15-25g)
- No significant impact of the sex on treatment efficacy over a similar data sample between males and females

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_combined=study_results.merge(mouse_metadata,on="Mouse ID")
# mouse_combined=pd.merge(study_results,mouse_metadata,on="Mouse ID").
# Display the data table for preview
mouse_combined.head()
```

```python
# Checking the number of mice.
len(mouse_combined["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_combined[mouse_combined.duplicated(["Mouse ID", "Timepoint"])]["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_combined.loc[mouse_combined["Mouse ID"]==mouse_combined[mouse_combined.duplicated(["Mouse ID", "Timepoint"])]["Mouse ID"].unique()[0],:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_clean=mouse_combined.set_index(["Mouse ID"])
mouse_clean.drop(mouse_combined[mouse_combined.duplicated(["Mouse ID", "Timepoint"])]["Mouse ID"].unique()[0],inplace=True)
mouse_clean.reset_index(inplace=True)
mouse_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(mouse_clean["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mouse_summary=pd.DataFrame({"Mean Tumor Volume":mouse_clean.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean(),
                            "Median Tumor Volume":mouse_clean.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median(),
                            "Tumor Volume Variance":mouse_clean.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var(),
                            "Tumor Volume Std. Dev.":mouse_clean.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std(),
                            "Tumor Volume Std. Err.":mouse_clean.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
                           })
mouse_summary



# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
mouse_clean.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].agg(["mean", "median","var","std","sem"])
# Using the aggregation method, produce the same summary statistics in a single line
```

## Correlation and Regression

```python
plt.scatter(mouse_weight.groupby("Mouse ID")["Weight (g)"].unique(),mouse_weight.groupby("Mouse ID")["Tumor Volume (mm3)"].mean())
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

x=mouse_weight.groupby("Mouse ID")["Weight (g)"].unique().astype(float)
y=mouse_weight.groupby("Mouse ID")["Tumor Volume (mm3)"].mean().tolist()
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x, y)
plt.plot(x,x*slope+intercept,'r')
plt.show()
```
